```python
import numpy as np
import MonotonicSpline as ms
import matplotlib.pyplot as plt
from scipy.interpolate import CubicSpline
```

```python
x_array = np.linspace(1,30)
y_array = 15*np.random.rand(len(x_array))
```

```python
plt.scatter(x_array, y_array)
plt.show()
```

```python
cspline_mine = ms.Cubic(x_array, y_array)
cspline_scipy = CubicSpline(x_array, y_array)
```

```python
x_vals = np.linspace(1, 30, 500)
```

```python
plt.plot(x_vals, cspline_scipy(x_vals), label='scipy')
plt.plot(x_vals, cspline_mine.predict(x_vals), label='mine')
plt.legend(loc=1)
plt.show()
```

```python
plt.close()
y_array = 500*np.random.rand(len(x_array))
cspline_mine = ms.Cubic(x_array, y_array)
cspline_scipy = CubicSpline(x_array, y_array)
```
